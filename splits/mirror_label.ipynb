```python
import re

# 1. 유효한 범위 정의
valid_ranges = {
    "drive_0004": [(240, 358)],
    "drive_0005": [(130, 379)],
    "drive_0006": [(357, 522)],
    "drive_0007": [(500, 730)],
    "drive_0008": [(139, 321)],
    "drive_0009": [(251, 349)],
    "drive_0010": [(170, 367)],
    "drive_0011": [(507, 610)],
    "drive_0012": [(72, 253)],
    "drive_0013": [(62, 121)],
}

# 2. 프레임이 유효한지 확인하는 함수
def is_valid_frame(drive_name, frame_num):
    if drive_name not in valid_ranges:
        return False
    return any(start <= frame_num <= end for start, end in valid_ranges[drive_name])

# 3. 파일 처리 함수
def process_txt_file(input_path, output_path):
    with open(input_path, 'r') as f:
        lines = f.readlines()

    result_lines = []
    for line in lines:
        line = line.strip()
        if not line:
            continue

        try:
            path, frame_str, r = line.split()
            frame = int(frame_str)
            match = re.search(r"(drive_\d+)_sync", path)
            if not match:
                continue
            drive = match.group(1)
            label = 1 if is_valid_frame(drive, frame) else 0
            result_lines.append(f"{path} {frame} {r} {label}")
        except ValueError:
            print(f"라인 처리 오류: {line}")
            continue

    # 4. 출력 파일 저장
    with open(output_path, 'w') as f:
        for line in result_lines:
            f.write(line + '\n')

    print(f"✅ 처리 완료: {len(result_lines)}줄 저장됨 → {output_path}")

# 5. 경로 지정
input_txt_path = "./jbnu_stereo/all_files.txt"   # 입력 파일 이름
output_txt_path = "./jbnu_stereo/mirror_files.txt" # 출력 파일 이름

# 실행
process_txt_file(input_txt_path, output_txt_path)
```
